## Plot genome coverage for each assembly as determined by MUMmer and MASMVALI

```python
import os

for i in [1, 2, '2b']:
    for j in [31, 41 ,51]:
        f = "NewMock{i}/out_{j}/mummer/metagenome-coverage.1000.tsv".format(i=i, j=j)
        if os.path.isfile(f):
            mgcovi = pd.read_csv(f, sep="\t").sort("genome").reset_index(drop=True)
            mgcovi.plot(kind="bar", y="contig_cov_percentage", title=f)
            show()
```

## After removing duplicate contigs

```python
ccount_new = pd.read_csv("NewMock1/fprau-bug/out_41/remap-nodup/contig_count_per_genome_4f4685e.tsv", sep="\t").sort("contig").reset_index(drop=True)
b = pd.concat([cinfo1.ix[:, 1], ccount_new.ix[:, 1:21].idxmax(axis=1)], axis=1)
b.columns=["length", "genome"]
b.groupby("genome").sum().sort('length')
```

```python
bb = b.groupby("genome").sum().sort()
bb.reset_index(inplace=True)
bb.sort("genome")
c = pd.concat([bb.sort("genome"), mgcov1.ix[0:19, 0:2]], axis=1)
c
c.columns = ["genome1", "sum_contigs", "genome2", "real_length"]
c["ratio"] = c.sum_contigs.div(c.real_length)
c[["genome1","ratio","sum_contigs","real_length"]].sort("ratio")
#c.sort("sum_contigs")
```

## Taking only contigs >= 1K

```python
ccount_new = pd.read_csv("NewMock1/fprau-bug/out_41/remap-nodup/contig_count_per_genome_4f4685e.tsv", sep="\t").sort("contig").reset_index(drop=True)
b = pd.concat([cinfo1.ix[:, 1], ccount_new.ix[:, 1:21].idxmax(axis=1)], axis=1)
b.columns=["length", "genome"]
bb = b.loc[b.length >= 1000].groupby("genome").sum().sort()
bb.reset_index(inplace=True)
bb.sort("genome")
c = pd.concat([bb.sort("genome"), mgcov1.ix[0:19, 0:2]], axis=1)
c
c.columns = ["genome1", "sum_contigs", "genome2", "real_length"]
c["ratio"] = c.sum_contigs.div(c.real_length)
c[["genome1","ratio","sum_contigs","real_length"]].sort("ratio")
```
