Divide into copy number variation groups of
   
   - no gain vs gain ``[-2,1,0]`` and ``[1,2]``
   - loss vs no loss ``[-2,-1]`` and ``[0,1,2]``

```python
def split_groups(t, threshold):
    """Divide row t in two based on copy number. Perform Mann-Witney U test. Return p value"""
    # check if expression data exists
    if t.name in exp.index:
        x = exp.ix[t.name, t[t < threshold].index].dropna()
        y = exp.ix[t.name, t[t >= threshold].index].dropna()
    else:
        return 1

    # check group size
    if min(len(x), len(y)) < 0.10 * max(len(x), len(y)):
        return 1
    # scipy doc says Mann-Witney U should have more than 20 samples in x and y
    elif min(len(x), len(y)) <= 20:
        return 1
    # check if there is actual expression
    elif x.sum() + y.sum() < 1:
        return 1
    # check that the direction of the expression is similar to CNV
    elif x.mean() >= y.mean():
        return 1
    else:
        # perform Mann-Witney U-test, also call Wilcoxon rank sum test (not to be confused with Wilcoxon signed rank test)
        return scipy.stats.mannwhitneyu(x, y)[1]
```

```python
# Corner cases split_groups for testing
# non-existent gene expression data
e1 = cn.ix["FAM138F|641702|1of3", :]
# all expression is 0
e2 = cn.ix["PRAMEF3", :]
```

```python
cn_loss_pvalue = cn.ix[:,2:len(cn.columns)].apply(split_groups, axis=1, args=(0,))
```

```python
cn_gain_pvalue = cn.ix[:,2:len(cn.columns)].apply(split_groups, axis=1, args=(1,))
```

```python
from statsmodels.sandbox.stats.multicomp import multipletests
```

## Per subtype

```python
tcga = pd.read_csv("MyTCGA_table.txt", sep="\t")
```

```python
tcga.groupby("PAM50").Sample.count()
```

```python
subtypes = {g[0]:g[1].Sample.apply(lambda x: x[:-1]) for g in tcga.groupby("PAM50")}
```

```python
%%time
# create data frame to hold p values per subtype
cn_reg = pd.DataFrame(index=cn.index)
for k in subtypes:
    # loss
    cn_reg[k + "_loss_pvalue"] = cn.ix[:,subtypes[k]].apply(split_groups, axis=1, args=(0,))
    bh_res = multipletests(cn_reg[k + "_loss_pvalue"], alpha=0.05, method="fdr_bh")
    cn_reg[k + "_loss_adj_bhr_pvalue"] = bh_res[1]
    cn_reg[k + "_loss_adj_bhr_pvalue_bool"] = bh_res[0]
    # gain
    cn_reg[k + "_gain_pvalue"] = cn.ix[:,subtypes[k]].apply(split_groups, axis=1, args=(1,))
    bh_res = multipletests(cn_reg[k + "_gain_pvalue"], alpha=0.05, method="fdr_bh")
    cn_reg[k + "_gain_adj_bhr_pvalue"] = bh_res[1]
    cn_reg[k + "_gain_adj_bhr_pvalue_bool"] = bh_res[0]
    # amplification
    cn_reg[k + "_amp_pvalue"] = cn.ix[:,subtypes[k]].apply(split_groups, axis=1, args=(2,))
    bh_res = multipletests(cn_reg[k + "_amp_pvalue"], alpha=0.05, method="fdr_bh")
    cn_reg[k + "_amp_adj_bhr_pvalue"] = bh_res[1]
    cn_reg[k + "_amp_adj_bhr_pvalue_bool"] = bh_res[0]
```

```python
cn_reg.head()
```

```python
for k in subtypes:
    v = """
    Stats for {}
    Number of samples: {}
    Number of genes copy number regulated with adj p < 0.05
    GAIN: {}
    LOSS: {}
    AMP: {}
    UNION: {}
    INTERSECT: {}
    """.format(k,
               len(subtypes[k]),
               (cn_reg[k + "_gain_adj_bhr_pvalue_bool"]).sum(),
               (cn_reg[k + "_loss_adj_bhr_pvalue_bool"]).sum(),
               (cn_reg[k + "_amp_adj_bhr_pvalue_bool"]).sum(),
               (cn_reg[k + "_gain_adj_bhr_pvalue_bool"] | cn_reg[k + "_loss_adj_bhr_pvalue_bool"] | cn_reg[k + "_amp_adj_bhr_pvalue_bool"]).sum(),
               (cn_reg[k + "_gain_adj_bhr_pvalue_bool"] & cn_reg[k + "_loss_adj_bhr_pvalue_bool"] & cn_reg[k + "_amp_adj_bhr_pvalue_bool"]).sum(),
               )
    print v
```

```python
all_inter = pd.Series([True,]*len(cn_reg))
all_union = pd.Series([False,]*len(cn_reg))

for g in ("gain", "loss", "amp"):
    for k in subtypes:
        all_inter = np.bitwise_and(all_inter, cn_reg["{}_{}_adj_bhr_pvalue_bool".format(k, g)])
        all_union = np.bitwise_or(all_union, cn_reg["{}_{}_adj_bhr_pvalue_bool".format(k, g)])
v = """
All subtypes
Number of genes copy number regulated with adj p < 0.05
UNION: {}
INTERSECT: {}
""".format(all_union.sum(),
           all_inter.sum())
print v
```

This number is different from the initial 13K when not subtyping, prolly because of the way the adjusted p values have been computed. Should think about whether to include all subtypes when doing the Benjamini Hochberg procedure or per subtype.

```python
def get_cna_seg_mean_per_sample(row, cnv):
    return cnv[(cnv.Chromosome == str(row.chrom)[3:]) & (cnv.Start <= row.txStart) & (cnv.End >= row.txEnd)] \
        .groupby("Sample")["Segment_Mean"].median()
```

```python
def spearmanr_check(x, exp):
    spearr, p2tailed = scipy.stats.spearmanr(x, exp.ix[x.name, :])
    return {"spearr":float(spearr), "p2tailed":float(p2tailed)}
```
