# Find Genes that are Copy Number Regulated using TCGA

```python
import pandas as pd
import numpy as np
import scipy.stats
```

- Copy number data per gene

```python
cn = pd.read_csv("copynumber.txt", sep="\t").set_index("Gene Symbol")
```

```python
cn.head()
```

- expression data per gene

```python
# get expression column to sample mapping
file_sample_map = {".".join(l.split("\t")[0].split(".")[:4]):l.split("\t")[1][:-13] for l in [r.strip() for r in open("FILE_SAMPLE_MAP.txt").readlines()][1:-1]}
# get gene locus to gene id mapping
locus_gene_id_map = {cn.ix[i,"Locus ID"]:i for i in cn.index}
locus_gene_id_map_func = lambda x: locus_gene_id_map.get(int(x.split("|")[1]), "NOT_IN_CN")
exp = pd.read_csv("expression.txt", sep="\t", converters={"gene_id": locus_gene_id_map_func}) \
    .rename(columns=file_sample_map).set_index("gene_id")
```

```python
exp.head()
```

Divide into copy number variation groups of
   
   - no gain vs gain ``[-2,1,0]`` and ``[1,2]``
   - loss vs no loss ``[-2,-1]`` and ``[0,1,2]``

```python
def split_groups(t, threshold):
    """Divide row t in two based on copy number. Perform Mann-Witney U test. Return p value"""
    # check if expression data exists
    if t.name in exp.index:
        x = exp.ix[t.name, t[t < threshold].index].dropna()
        y = exp.ix[t.name, t[t >= threshold].index].dropna()
    else:
        return 1

    # check group size
    if min(len(x), len(y)) < 0.10 * max(len(x), len(y)):
        return 1
    # check if there is actual expression
    elif x.sum() + y.sum() < 1:
        return 1
    else:
        # perform Mann-Witney U-test, also call Wilcoxon rank sum test (not to be confused with Wilcoxon signed rank test)
        return scipy.stats.mannwhitneyu(x, y)[1]
```

```python
# Corner cases split_groups for testing
# non-existent gene expression data
e1 = cn.ix["FAM138F|641702|1of3", :]
# all expression is 0
e2 = cn.ix["PRAMEF3", :]
```

```python
cn_loss_pvalue = cn.ix[:,2:len(cn.columns)].apply(split_groups, axis=1, args=(0,))
```

```python
cn_gain_pvalue = cn.ix[:,2:len(cn.columns)].apply(split_groups, axis=1, args=(1,))
```

```python
from statsmodels.sandbox.stats.multicomp import multipletests
```

- Adjust p value with Benjamini Hochberg

```python
cn_loss_adj_pvalue = multipletests(cn_loss_pvalue, alpha=0.05, method="fdr_bh")
cn_gain_adj_pvalue = multipletests(cn_gain_pvalue, alpha=0.05, method="fdr_bh")
```
