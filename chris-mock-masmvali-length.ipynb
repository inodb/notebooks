Name sets of names, so we can compare each set separately. Otherwise you don't see anything in the plot

```python
asm_sets = {}asm_sets['contig'] = [n for n in names if n.endswith('noscaf')]asm_sets['scaf'] = [n for n in names if n.endswith('scaf') and not n.endswith('noscaf')]asm_sets['merge'] = [n for n in names if n.endswith('newbler') or n.endswith('minimus2')]asm_sets
```

Plot length distribution function, uses masmvali

```python
def plot_length(assemblies, names, colors, linestyles, cut_off=100):    ax = subplot(1,1,1)    for i, asm in enumerate(assemblies):        contigs = assembly.ContigDict()        contigs.parse_fasta_lengths(asm)        contiglens = [c.length for c in contigs.itervalues()]        contiglens.sort()        contiglens.reverse()        cla = np.array(contiglens)        plot(cla[cla > cut_off], np.cumsum(cla[cla > cut_off]), color=colors[i], label=names[i], linestyle=linestyles[i], linewidth=2.0)    plt.gca().invert_xaxis()    # Labels    plt.xlabel('Contig/Scaffold length')    plt.ylabel('Sum of bases')    # Legend    handles, labels = ax.get_legend_handles_labels()    ax.legend(handles, labels)
```

Compare all sets

```python
# Change figure sizercParams['figure.figsize'] = 10, 10for k in asm_sets:    print "Comparing: " + k    plot_length([filesd[name] for name in asm_sets[k]], asm_sets[k], [colord[name] for name in asm_sets[k]], [shapesd[name] for name in asm_sets[k]])    plt.show()
```
