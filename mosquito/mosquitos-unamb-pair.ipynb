## Checking unambiguous read counts
### Mapping reads with bowtie2
First mapped the reads with bowtie2 and removed duplicates:
```
cd /proj/b2013255/private/nobackup/mapping-project/inod/map/1000LCO_S2
for s in 1000LCO_S2_L001_R1_001.fastq
do
    ls bowtie2/asm_pair-smds.coverage || 
    sbatch --output=bowtie2-slurm.out -A b2010008 -t 06:00:00 -J mosquito -p core -n 4 ~/bin/sbatch_job \
    bash /glob/inod/github/metassemble/scripts/map/map-bowtie2-markduplicates.sh -ct 4 $s ${s/R1/R2} \
    pair ../swedish_mosquitoes_contigs_flat.fasta ref bowtie2
done
```
Took about ~8 min to run. Then checked for unambiguously aligning pairs. They don't have the XS tag set and they should be properly aligning e.g. 0x0002 according to the [sam spec](http://samtools.github.io/hts-specs/SAMv1.pdf).
```
cd /proj/b2013255/private/nobackup/mapping-project/inod/map/1000LCO_S2/bowtie2/unambiguous-aligning-pairs
samtools view ../ref_pair-smds.bam | \
    awk '{if ($13 !~ "XS" && and($2, 0x0002) && $7 == "=") {print $0}}' | 
    samtools view -bT ../../../swedish_mosquitoes_contigs_flat.fasta - > ref_pair-smds.bam
genomeCoverageBed -ibam ref_pair-smds.bam > ref_unamb_pair.coverage
python $METASSEMBLE_DIR/scripts/validate/map/gen_contig_cov_per_bam_table.py --isbedfiles 
    ../../../swedish_mosquitoes_contigs_flat.fasta ref_unamb_pair.coverage > ref_unamb_pair.coverage.percontig
```
The scripts to parse the BED histogram and do the mapping are from [metassemble](https://github.com/inodb/metassemble)

### Plot the coverages and percentages covered

````python
%matplotlib inline

mpl.rcParams['figure.figsize'] = 8, 6

cov.ix[:,['cov_mean_sample_0','percentage_covered_sample_0']].plot(kind='bar', legend=False, secondary_y=['percentage_covered_sample_0'],)
````

So we can see that there's one genome that is covered way more than the other ones. After inspecting the bam file in [Tablet](http://ics.hutton.ac.uk/tablet/) it would seem that there are quite a few mismatches for *A. beklemishevi*. Not sure if this is some mapping problem. *A. Punctodes* is only 268bp so it is not that strange that we don't have any pairs aliging properly to this reference. Taking only reads with zero mismatches or 1:

```
samtools view ../ref_pair-smds.bam | 
    awk '{if ($13 !~ "XS" && and($2, 0x0002) && $7 == "=" && ($0 ~ "XM:i:0        " || $0 ~ "XM:i:1       ")) {print $0}}' |
    samtools view -bT ../../../swedish_mosquitoes_contigs_flat.fasta - > ref_unamb_1mismatch_pair.bam
```

Basically only *C. Morsitans* gets matches then:

````python
cov_unamb = pd.read_csv("ref_unamb_1mismatch_pair.coverage.percontig", sep='\t').sort('contig').set_index('contig')
%matplotlib inline
mpl.rcParams['figure.figsize'] = 8, 6
cov_unamb.ix[:,['cov_mean_sample_0','percentage_covered_sample_0']].plot(kind='bar', legend=False, secondary_y=['percentage_covered_sample_0'],)
````

It depends on the sample whether this approach can be called successfull or not. If only *C. Morsitans* was in there, it is extremely effective.
